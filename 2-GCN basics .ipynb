# Graph Laplacian
- Laplacian Operator: 
$$\Delta: L^2(\mathcal{V}) \rightarrow L^2(\mathcal{V})$$
$$\Delta(f_i)=\frac{1}{b_i} \sum_{j:(i,j)}a_ia_j(f_i-f_j)$$
- Intuitively, it measures the divergence and is the difference between f and its local average.
- It is analogous to second order derivative on graphs.
- Smoothness of a function results in lower values for the Laplacian applied to the function.

# Graph Laplacian Variations
- Unnormalized: $\Delta = D - A$
- Normalized: $\Delta = I-D^{-\frac{1}{2}}AD^{-\frac{1}{2}}$
- Random Walk: $\Delta = I - D^{-1}A$

# Spectral Decomposition
- A Laplacian of a graph with n vectors admits n eigenvectors. $\Delta \phi_i = \lambda_i \phi_k, k=1,2,...$
- Laplacian eigenvectors are also called Fourier basis function/modes
- Eigendecomposition of a graph Laplacian: $\Delta = \Phi^T \Lambda \Phi$
- Smoothness is proportional to local invariance. Reducing Laplacian results in smoothness and thus local invariance. We would need to minimize the Laplacian and the solution is given by the eigenvector.


# Fourier Analysis on Graphs
- Laplacian eigenvectors (Fourier modes) are related to topology of the graph.
- The first eigenvectors identify communities on the graph.
- Spectral Clustering: identifying communities by computing first $k$ eigenvectors


# Fourier Decomposition for Graphs
- Convolution is easier on Fourier modes
- $\hat{f} = \Phi^T f$, $\phi_i$ is Graph Fourier Basis or Lap eigenvectors$

$$
\Delta (\phi)k=\lambda_k \phi_k \text{ with }
\begin{cases}
\phi_k = \text{graph Fourier mode} \\
\lambda_k = square (frequency)
\end{cases}
$$

# Convolution on Graphs
- Convolutions consists of convolutions and downsampling
- Convolution theorem: $\widehat{f*g}=\hat{f}.\hat{g}$
- Convolution on graphs is applying Laplacian to $f$ with some spectral function. $$f*g=\hat{g}(\Delta)f$$

# Downsampling in Graph
- The goal is to create a pooling or some other sort of aggregation across multiple layers in order to create global invariance.
- The coarsening operator must preserve non-linearity.
- Graph coarsening is same partitioning graph to meaningful communities.
